```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with IntelliTech Squad [Final Project AOT 5th to 7th Sem CSBS (23-24)] Data Mavericks - ML Innovators.txt'
,'r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
message = re.split(pattern, data)[1:]
message
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':message, 'message_date':dates})

#convert date_type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M - ')
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
#Separate users and message

# Assuming df is your DataFrame and it has a column named 'user_message'
user = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        user.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        user.append('group_notification')
        messages.append(entry[0])

df['user'] = user
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```
